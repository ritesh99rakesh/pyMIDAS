# MIDAS-R example

## Import packages

```python
from midas import midasR
from auc import auc
import pandas as pd
import numpy as np
```

## Load data

```python
data = pd.read_csv(
    'darpa_processed_data.csv',
    names=['src', 'dst', 'times'],
)
```

## Find the anomaly score

```python
anomaly_score = midasR(
    data.src, data.dst, data.times,
    num_rows=2,
    num_buckets=769,
    factor=0.4,
)
```

## Save the anomaly score for finding AUC

```python
np.savetxt('scores.txt', anomaly_score)
```

## Find the AUC score

```python
auc()
```
